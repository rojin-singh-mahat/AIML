```python
def convert_units(conversion_type, value):
    """Convert between supported length, weight, and volume units."""
    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


def run_unit_converter_program():
    print("Choose a conversion type:")
    print("1. meters to feet")
    print("2. feet to meters")
    print("3. kilograms to pounds")
    print("4. pounds to kilograms")
    print("5. liters to gallons")
    print("6. gallons to liters")

    choice_map = {
        "1": "m_to_ft",
        "2": "ft_to_m",
        "3": "kg_to_lbs",
        "4": "lbs_to_kg",
        "5": "l_to_gal",
        "6": "gal_to_l"
    }

    choice = input("Enter your choice (1-6): ")

    try:
        if choice not in choice_map:
            raise ValueError("Invalid conversion choice.")

        value = float(input("Enter the value to convert: "))
        result = convert_units(choice_map[choice], value)
        print("Converted value:", result)
    except ValueError as error:
        print("Error:", error)


print("Sample conversions:")
print("5 meters to feet =", convert_units("m_to_ft", 5))
print("10 kilograms to pounds =", convert_units("kg_to_lbs", 10))
print("3 liters to gallons =", convert_units("l_to_gal", 3))
```

### Qn 2

```python
def find_sum(numbers):
    """Return sum of all numbers in the list."""
    total = 0
    for number in numbers:
        total += number
    return total


def find_average(numbers):
    """Return average of numbers in the list."""
    if len(numbers) == 0:
        raise ValueError("The list is empty.")
    return find_sum(numbers) / len(numbers)


def find_maximum(numbers):
    """Return maximum value in the list."""
    if len(numbers) == 0:
        raise ValueError("The list is empty.")

    maximum = numbers[0]
    for number in numbers:
        if number > maximum:
            maximum = number
    return maximum


def find_minimum(numbers):
    """Return minimum value in the list."""
    if len(numbers) == 0:
        raise ValueError("The list is empty.")

    minimum = numbers[0]
    for number in numbers:
        if number < minimum:
            minimum = number
    return minimum


def run_math_program():
    print("Choose an operation:")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    choice = input("Enter your choice (1-4): ")

    try:
        raw_numbers = input("Enter numbers separated by spaces: ")
        parts = raw_numbers.split()

        if len(parts) == 0:
            raise ValueError("You entered an empty list.")

        numbers = []
        for part in parts:
            numbers.append(float(part))

        if choice == "1":
            print("Sum =", find_sum(numbers))
        elif choice == "2":
            print("Average =", find_average(numbers))
        elif choice == "3":
            print("Maximum =", find_maximum(numbers))
        elif choice == "4":
            print("Minimum =", find_minimum(numbers))
        else:
            print("Invalid operation choice.")
    except ValueError as error:
        print("Error:", error)


sample_numbers = [4, 8, 15, 16, 23, 42]
print("Sample numbers:", sample_numbers)
print("Sum =", find_sum(sample_numbers))
print("Average =", find_average(sample_numbers))
print("Maximum =", find_maximum(sample_numbers))
print("Minimum =", find_minimum(sample_numbers))
```

```python
def extract_every_other(lst):
    return lst[::2]


def get_sublist(lst, start, end):
    return lst[start:end + 1]


def reverse_list(lst):
    return lst[::-1]


def remove_first_last(lst):
    return lst[1:-1]


def get_first_n(lst, n):
    return lst[:n]


def get_last_n(lst, n):
    return lst[-n:]


def reverse_skip(lst):
    return lst[-2::-2]


example_list = [1, 2, 3, 4, 5, 6]
print("Original list:", example_list)
print("1. Every other element:", extract_every_other(example_list))
print("2. Sublist from index 2 to 4:", get_sublist(example_list, 2, 4))
print("3. Reversed list:", reverse_list([1, 2, 3, 4, 5]))
print("4. Remove first and last:", remove_first_last([1, 2, 3, 4, 5]))
print("5. First 3 elements:", get_first_n([1, 2, 3, 4, 5], 3))
print("6. Last 2 elements:", get_last_n([1, 2, 3, 4, 5], 2))
print("7. Reverse skip:", reverse_skip(example_list))
```

### Qn 4

```python
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            for value in flatten(item):
                result.append(value)
        else:
            result.append(item)
    return result


def access_nested_element(lst, indices):
    current = lst
    for index in indices:
        current = current[index]
    return current


def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


def remove_element(lst, elem):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        else:
            if item != elem:
                result.append(item)
    return result


def find_max(lst):
    flat_list = flatten(lst)
    maximum = flat_list[0]
    for item in flat_list:
        if item > maximum:
            maximum = item
    return maximum


def count_occurrences(lst, elem):
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count


def deep_flatten(lst):
    return flatten(lst)


def average_nested(lst):
    flat_list = flatten(lst)
    if len(flat_list) == 0:
        raise ValueError("Nested list is empty.")
    return sum(flat_list) / len(flat_list)


nested_example_1 = [[1, 2], [3, 4], [5]]
nested_example_2 = [[1, 2], [3, [4, 5]], 6]
nested_example_3 = [[1, 2], [3, 2], [4, 5]]
nested_example_4 = [[[1, 2], [3, 4]], [[5, 6], [7, 8]]]

print("1. Flatten:", flatten(nested_example_1))
print("2. Access nested element:", access_nested_element([[1, 2, 3], [4, 5, 6], [7, 8, 9]], [1, 2]))
print("3. Sum nested:", sum_nested(nested_example_2))
print("4. Remove element 2:", remove_element(nested_example_3, 2))
print("5. Find max:", find_max(nested_example_2))
print("6. Count occurrences of 2:", count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
print("7. Deep flatten:", deep_flatten(nested_example_4))
print("8. Average nested:", average_nested([[1, 2], [3, 4], [5, 6]]))
```

```python
size = 1_000_000

list_a = list(range(size))
list_b = list(range(size))
array_a = np.array(list_a)
array_b = np.array(list_b)

start = time.time()
list_add_result = []
for i in range(size):
    list_add_result.append(list_a[i] + list_b[i])
list_add_time = time.time() - start
print("Python list addition time:", list_add_time, "seconds")

start = time.time()
array_add_result = array_a + array_b
array_add_time = time.time() - start
print("NumPy array addition time:", array_add_time, "seconds")

start = time.time()
list_mul_result = []
for i in range(size):
    list_mul_result.append(list_a[i] * list_b[i])
list_mul_time = time.time() - start
print("Python list multiplication time:", list_mul_time, "seconds")

start = time.time()
array_mul_result = array_a * array_b
array_mul_time = time.time() - start
print("NumPy array multiplication time:", array_mul_time, "seconds")

start = time.time()
list_dot_result = 0
for i in range(size):
    list_dot_result += list_a[i] * list_b[i]
list_dot_time = time.time() - start
print("Python list dot product time:", list_dot_time, "seconds")

start = time.time()
array_dot_result = np.dot(array_a, array_b)
array_dot_time = time.time() - start
print("NumPy array dot product time:", array_dot_time, "seconds")

print("\nSummary for the first three tests:")
print("Addition -> Python:", list_add_time, "NumPy:", array_add_time)
print("Multiplication -> Python:", list_mul_time, "NumPy:", array_mul_time)
print("Dot product -> Python:", list_dot_time, "NumPy:", array_dot_time)

run_full_matrix_test = False
matrix_size = 1000

if run_full_matrix_test:
    python_matrix_1 = []
    python_matrix_2 = []
    for i in range(matrix_size):
        row1 = []
        row2 = []
        for j in range(matrix_size):
            row1.append(i + j)
            row2.append(i - j)
        python_matrix_1.append(row1)
        python_matrix_2.append(row2)

    start = time.time()
    python_matrix_result = []
    for i in range(matrix_size):
        result_row = []
        for j in range(matrix_size):
            total = 0
            for k in range(matrix_size):
                total += python_matrix_1[i][k] * python_matrix_2[k][j]
            result_row.append(total)
        python_matrix_result.append(result_row)
    python_matrix_time = time.time() - start
    print("Python list matrix multiplication time:", python_matrix_time, "seconds")

    numpy_matrix_1 = np.array(python_matrix_1)
    numpy_matrix_2 = np.array(python_matrix_2)

    start = time.time()
    numpy_matrix_result = np.dot(numpy_matrix_1, numpy_matrix_2)
    numpy_matrix_time = time.time() - start
    print("NumPy matrix multiplication time:", numpy_matrix_time, "seconds")
else:
    print("\n1000x1000 matrix multiplication block skipped.")
```
